**Grouping Data:**
This method is used to group the data based on particular column
- It is called as "groupby"
- It is mainly used when we want to perform mathematical operations(Average..).
- function:---->  df.groupby('column_name')

```python
#Grouping data
df.groupby('Gender').size()
```

```python
df.groupby('Gender')['LoanAmount'].mean()  #calculating the mean loan amount for each gender
```

**Feature Engineering:**
This method is used to create or Engineer a new data based on exsisting data.
- Example is creating new column based on exsisting data.

```python
#Feature engineering
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']  #creating a new column 'TotalIncome' by adding 'ApplicantIncome' and 'CoapplicantIncome'
df.head()
```

```python
df['Valuable'] = 'No'
df.loc[(df['ApplicantIncome']> 4000) & (df['LoanAmount']> 200), 'Valuable'] = 'Yes'  #creating a new column 'Valuable' and assigning 'Yes' to rows where 'ApplicantIncome' is greater than 4000 and 'LoanAmount' is greater than 200, otherwise assigning 'No'
df.tail()
```

**Pivot Table:**
This method is used for transforming data to make Data Analysis.
- This concept has three important attributes such as: index , columns , values.
- In values it should be numerical.
- function:-----> df.pivot_table(index = ['column_name'],column = ['column_name'], values = ['column_name'])
                      

```python
#pivot tables
df.pivot_table(index = 'Gender', columns = 'Married', values = 'LoanAmount')  #creating a pivot table to count the number of occurrences of each education level for each combination
```

```python
df.pivot_table(index = 'Education', columns = 'Gender', values = 'TotalIncome')
```

**Merging Dataframes:**
This is used to merge two dataframes into one
- To do merging there should be common cloumn in both dataframes.
- "innerjoin" is used to display common rows (intersection).
- "outerjoin" is used to display all (Union).
- function:------>  pd.merge(df1,df2 , on='common_column', how ='inner/outer')

```python
#dataframe1
Students = pd.DataFrame({
       'student_id': [1, 2, 3, 4, 5],
       'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'age': [20, 21, 19, 22, 20]
             })
Students.head()
```

```python
#dataframe2
Courses = pd.DataFrame({
       'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
       'course_name': ['Mathematics', 'Physics', 'Chemistry', 'Biology', 'Computer Science', 'Art'],
       'credits': [3, 4, 3, 4, 3, 2]
             })
Courses
```

```python
#merging dataframes
merged_df = pd.merge(Students, Courses, on = 'name')  #merging the Students and Courses DataFrames on the 'name' column
merged_df
```

```python
#merging dataframes based on inner join
merged_df = pd.merge(Students, Courses, on = 'name', how='inner')
merged_df
```

```python
#merging dataframes based on outer join
merged_df = pd.merge(Students, Courses, on = 'name', how='outer')
merged_df
```
